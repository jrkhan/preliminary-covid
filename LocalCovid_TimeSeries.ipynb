```
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```
url = 'https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv'
df1 = pd.read_csv(url)
```

```
monroeFilter = df1.county == "Monroe"
onodagaFilter = df1.county == "Onondaga"
stateFilter = df1.state == "New York"
```

```
def prepCounty(countyFilter, stateFilter):
    county = df1[countyFilter & stateFilter]
    tdf = pd.DataFrame()
    tdf['county'] = county.county;
    tdf['dailyCases'] = county.cases.diff()
    tdf['date'] = county.date
    tdf.date = pd.to_datetime(tdf.date);
    tdf.reindex(tdf.date)
    return tdf
```

```
monroedf = prepCounty(monroeFilter, stateFilter)
onondagadf = prepCounty(onodagaFilter, stateFilter)
```

```
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20, 10))
plt.plot('date', 'dailyCases', data=monroedf, color='blue', linewidth=3, label='Monroe County, NY')
plt.plot('date', 'dailyCases', data=onondagadf, color='orange', linewidth=3, label='Onondaga County, NY')
ax.xaxis.set_major_locator(mdates.WeekdayLocator())
plt.ylabel("Number of new confirmed cases each day")
plt.legend()
mpld3.display(fig)
```
